# Regression Test Problems

```python
from sklearn.datasets.samples_generator import make_regression, make_circles, make_blobs, make_moons
from matplotlib import pyplot as plt
%matplotlib inline
from pandas import DataFrame
```

```python
# Plot regression lines from randome samples
X, y = make_regression(n_samples=100, n_features=1, noise=0.5)
plt.scatter(X, y, color='red')
X2, y2 = make_regression(n_samples=50, n_features=1, noise=0.3)
plt.scatter(X2, y2, color='blue')
plt.show()
```

```python
# Plot multiple circles
X, y = make_circles(n_samples=100, noise = 0.05)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label = y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
# Plot blobs for classification problems
X, y = make_blobs(n_samples=100, centers=5, n_features=2)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green', 3:'black', 4:'orange'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', y='y', x='x', label=key, color=colors[key])
plt.show()
```

```python
# Plot moonms for nonlinear classifications 
X, y = make_moons(n_samples=100, noise=0.05)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0: 'green', 1: 'orange'}
fir, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', y='y', x='x', label=key, color=colors[key])
```
